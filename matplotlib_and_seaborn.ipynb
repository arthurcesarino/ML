# Plotting: Introduction to matplotlib

```matplotlib``` is a python library. It contains the ```pyplot``` module, wich is basically a collection of functions such as ```plot```,```title```, ```show()```, etc.

    
   


Starting for the basics:

# Basic Plotting, Axes Labels and Titles

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

#plotting two 1-D numpy arrays
x = np.linspace(5,100,100)
y = np.linspace(10,1000,100)

plt.plot(x,y)
````

````python
plt.show()
````

````python
#can also work with lists though it converts lists to np arrays
plt.plot([1,4,6,8],[3,8,3,5])
plt.show()
````

Lets see how to put labels and the x and y axes and the chart title

Also, you can specify the limits of x and y labels as a range using ```xlim([xmin, xmax])``` and ```ylim([ymin, ymax])```

````python
#Axis label and titles
plt.plot(x,y)

#x and y labels and title
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's law")

#define the range of labels of the axis
#Arguments: plt.axis(xmin,xmax,ymin,ymax)
plt.xlim([20,80])
plt.ylim([200,800])
plt.show()
    
````

````python
#change the colors and line type

#initialising x and y arrays
x = np.linspace(0,10,20)
y = x*2

#color red, line type '+'
plt.plot(x, y, 'r+')

#put x and y axes labels and the title
plt.xlabel("Current")
plt.ylabel("Voltage")
plt.title("Ohm's law")

plt.show()
````

````python
#plotting multiple lines on the same plot

x = np.linspace(0,5,10)
y = np.linspace(3,6,10)

#plot three curves: y, y**2 and y**3 with diferent line types
plt.plot(x, y, 'r-', x, y**2, 'b+', x, y**3, 'g^')

plt.show()
````

# Figures and subplots

You often need to create multiple plots in the same figure, as we'll see in some upcoming examples;

```matplotlib``` has the concept of figures and subplots using wich you can create multiple subplots inside the same figure.

To create multiple plots in the same figure, you can use the method ```plt.subplot(nrows,ncols, nsubplot)```.

````python
x = np.linspace(1,10,100)
y = np.log(x)

#initialize a new figure explicitly
plt.figure(1)

#create a subplot with 1 row and 2 columms

#create the first subplot in figure 1
plt.subplot(121)        #equivalent to plt.subplot(1, 2, 1)
plt.title("y = log(x)")
plt.plot(x, y)

#create the second subplot in figure 1
plt.subplot(122)        #equivalent to plt.subplot(1, 2, 2)
plt.title("y= log(x)**2")
plt.plot(x, y**2)

plt.show()
````

Another example: create 4 subplots in two rows and two columms

````python
#creating a figure having 4 subplots

x = np.linspace(1,10,100)

plt.figure(1)

#subplot 1
plt.subplot(2,2,1)
plt.title("Linear")
plt.plot(x,x)

#subplot 2
plt.subplot(2,2,2)
plt.title("Quadratic")
plt.plot(x,x**2)

#subplot 3
plt.figure(2)
plt.subplot(2,2,1)
plt.title("log")
plt.plot(x,np.log(x))

#subplot 4
plt.subplot(2,2,2)
plt.title("exponencial")
plt.plot(x, np.exp(x))

plt.show()
````

## Advanced plotting with SEABORN lib

```seaborn``` is a python library built on top of matplotlib. It creates much more attractive plots than matplotlib, and its often more concise than matplotlib when you want to costumize your plots, add collors, grinds, etc.

## Visualising Univariate Distributions

We have already visualised univariate distributions before using boxplot, histogram, etc. Let's now 
do that using ```seaborn```. We will use the sales data for the upcomming few exercises.

````python
import seaborn as sns

#set a seaborn style of your taste
sns.set_style("whitegrid")

#data
df = pd.read_csv("./global_sales_data/market_fact.csv")
````

## Histograms and Density Plots

Histograms and density plots show the frequency of a numeric variable along the y-axis, and the value along the x-axis. The ```sns.distplot()``` function plots a density curve, that this is aesthetically better then vanilla ```matplotlib```.

````python
#simple density plot
sns.distplot(df['Shipping_Cost'])
plt.show()
````

You can also plot what is known as the rug plot witch plots the actual data points as small vertical bars. The rug plot is simply specified as an argument of ```distplot()```.

````python
#rug = True
#plotting only a few points since rug takes a long while
sns.distplot(df['Shipping_Cost'][:200], rug=True)
plt.show()
````

Simple density plot(without the histogram bars) can be created by specifying ```hist=False```

````python
sns.distplot(df['Sales'], hist=False)
plt.show()
````

Since seaborn is using matplotlib behind the scenes, the usual matplotlib functions work well with seaborn. For example, you can use subplots to plot muntiple univariate distributions.

````python
#subplots

#Subplot 1 -> Sales
plt.figure(1)
plt.subplot(2,2,1)
plt.title('Sales')
sns.distplot(df['Sales'])

#subplot 2 -> Profit
plt.subplot(2,2,2)
plt.title('Profit')
sns.distplot(df['Profit'])

#subplot 3 -> Order_Quantity
plt.figure(2)
plt.subplot(2,2,1)
plt.title('Order_Quantity')
sns.distplot(df['Order_Quantity'])

#subplot 4 -> Shipping_Cost
plt.subplot(2,2,2)
plt.title('Shipping_Cost')
sns.distplot(df['Shipping_Cost'])

plt.show()
````

## Visualising Bivariate Distributions

Bivariate distributions are simple two univariate distributions plotted on x and y axes respectively
They help you observe the relationship between the two variables

They are also called joint distributions and are created using ```sns.jointplot()```

````python
#joint plot of profit and sales
sns.jointplot('Profit', 'Sales', df)
plt.show()

#same as sns.jointplot(df['Profit'],df['Sales'])
````

Notice that booth the distributions are heavlyskewed and all the points seem to be concentrated in on region. This is cause due to some extreme values of Profits and Sales which matplotlib is trying to accomodate in the limited space of the plot.

Lets remove that point and plot again:

````python
#removing the point and plotting again
df = df[(df.Profit < 10000) & (df.Sales < 20000)]

sns.jointplot('Sales', 'Profit', df)
plt.show()
````

You can adjust the arguments of the ```jointplot()``` to make the plot more readable. For e.g especifying ```kind=hex``` will create a 'hexbin plot'.

````python
#plotting low sales value orders
#hex plot
df = pd.read_csv('./global_sales_data/market_fact.csv')
df = df[(df.Profit < 100) & (df.Profit > -100) & (df.Sales < 200)]
sns.jointplot('Sales', 'Profit', df, kind='hex', color='k')
plt.show()
````

The bottom-right region represents orders where the Sales was high but the Profit was low, i.e even when the store is getting high revenue, the orders are still making losses. These are the kind of orders a business would avoid


## Plotting Pairwise Releationships

You'll find it helpful to plot pairwise releationship between multiple numeric variables. For e.g, here we have taken the price of some popular cryptocurrencies such as bitcoins, litecoins, ethereum, monero, neo, quantum and ripple.

Now, the crypto enthusiasts would know that the price of these currencies vary with each other. If bitcoin goes up, the others will likely follow suit, etc.

Now, say you want to trade in some currencies. Given a set of some cryptocurrencies, how will you decide when and which one to buy/sell? it will be helpfull to analyse  past data and indentify some trends in these currencies.

````python
#reading cryptocurrency files
btc = pd.read_csv('./crypto_data/bitcoin_price.csv')
ether = pd.read_csv('./crypto_data/ethereum_price.csv')
ltc = pd.read_csv('./crypto_data/litecoin_price.csv')
monero = pd.read_csv('./crypto_data/monero_price.csv')
neo = pd.read_csv('./crypto_data/neo_price.csv')
quantum = pd.read_csv('./crypto_data/qtum_price.csv')
ripple = pd.read_csv('./crypto_data/ripple_price.csv')

#putting a suffix with column names so that joins are easy
btc.columns = btc.columns.map(lambda x: str(x) + '_btc')
ether.columns = ether.columns.map(lambda x: str(x) + '_et')
ltc.columns = ltc.columns.map(lambda x: str(x) + '_ltc')
monero.columns = monero.columns.map(lambda x: str(x) + '_mon')
neo.columns = neo.columns.map(lambda x: str(x) + '_neo')
quantum.columns = quantum.columns.map(lambda x: str(x) + '_qt')
ripple.columns = ripple.columns.map(lambda x: str(x) + '_rip')

btc.head()
````

````python
#merging all the files by date
m1 = pd.merge(btc, ether, how='inner', left_on='Date_btc', right_on='Date_et')
m2 = pd.merge(m1, ltc, how='inner', left_on='Date_btc', right_on='Date_ltc')
m3 = pd.merge(m2, monero, how='inner', left_on='Date_btc', right_on='Date_mon')
m4 = pd.merge(m3, neo, how='inner', left_on='Date_btc', right_on='Date_neo')
m5 = pd.merge(m4, quantum , how='inner', left_on='Date_btc', right_on='Date_qt')

crypto = pd.merge(m5, ripple, how='inner', left_on='Date_btc', right_on='Date_rip')

crypto.head()


````

````python
#Subsetting only the closing prices column for plotting

curr = crypto[["Close_btc","Close_et","Close_ltc","Close_mon","Close_neo","Close_qt","Close_rip"]]

curr.head()
````

## Pairwise Scatter Plots

Now, since we have multiple numeric variables, ```sns.pairplot()``` is a good choice to plot all of them in one figure.

````python
#pairplot
sns.pairplot(curr)
plt.show()
````

````python
#You can also observe the correlation between the currencies 
#using df.corr()
cor = curr.corr()
round(cor,3)
````

The dataset above is a ``correlation matrix`` of cryptocurrencies.

## Heatmaps

It will be usefull and helpfull to visualise the correlation matrix itself using ``sns.heatmap()``.


````python
#figue size
plt.figure(figsize=(10,8))

#heatmap
sns.heatmap(cor, cmap='YlGnBu', annot=True)
plt.show()
````

## EXEMPLOS DE PLOTTING DIVERSOS


````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#set seaborn theme if you prefer
sns.set(style = 'white')

#read data
market_df = pd.read_csv('./global_sales_data/market_fact.csv')
customer_df = pd.read_csv('./global_sales_data/cust_dimen.csv')
product_df = pd.read_csv('./global_sales_data/prod_dimen.csv')
shipping_df = pd.read_csv('./global_sales_data/shipping_dimen.csv')
orders_df = pd.read_csv('./global_sales_data/orders_dimen.csv')
````

## Boxplots

We had created simple boxplots such as orders shown bellow. Now, let's plot multiple boxplots and see what they can tell uss the distribution of variables across categories

````python
#boxplot of a variable
sns.boxplot(y=market_df['Sales'])
plt.yscale('log')
plt.show()
````

Now, let's say you want to **compare the (distribution of) sales of various product categories**. Let's first merge the product data into the main dataframe.

````python
#merge a dataframe to add a categorical variable
df = pd.merge(market_df,product_df,how='inner',on='Prod_id')
df.head()
````

````python
#boxplot of a variable across various product categories
sns.boxplot(x='Product_Category',y='Sales', data=df)
plt.yscale('log')
plt.show()
````

So this tell us that the sales of office supplies are, on an average, lower than the other two categories. The sales of the tecnology and furniture categories seem much better. Note that each order can have multiple units of product sold, so Sales being higher/lower may be due to price per unit or the number of units.

Let's now plot the profit.

````python
#boxplot of a variable across various product categories
sns.boxplot(x='Product_Category',y='Profit',data=df)
plt.show()
````

Profit clearly has some outliers due to which the boxplots are unreadable. Let's remove some extreme values from profit and try plotting.

````python
df = df[(df.Profit < 1000) & (df.Profit > -1000)]

sns.boxplot(x='Product_Category',y='Profit',data=df)
plt.show()
````

You can see that though the category 'technology' has better sales number than other, it is also the one where the **most loss making transactions** happen. Going further:

````python
#adjust figure size
plt.figure(figsize=(10,8))

#subplot 1: Sales
plt.subplot(1,2,1)
sns.boxplot(x='Product_Category',y='Sales',data=df)
plt.yscale('log')
plt.title('Sales')

#subplot 2: Profit
plt.subplot(1,2,2)
sns.boxplot(x='Product_Category',y='Profit',data=df)
plt.title('Profit')
plt.show()
````

Now that we've compared Sales and Profit across product categories, let's drill down further and do the same across another categorical variabe: 'Costumer_Segment'

We'll need to add the costumer-related attributes (dimensions) to the dataframe.


````python
#merging with customer_df
df = pd.merge(df,customer_df,how='inner',on='Cust_id')
df.head()
````

````python
#boxplot of a variable across the customer segment
sns.boxplot(x='Customer_Segment',y='Profit',data=df)
plt.show()
````

You can **visualise the distribution across two categorical variables** using the ``hue=`` argument.

````python
#set figure_size for larger figure
plt.figure(num=None, figsize=(12,8), dpi=80, facecolor='w', edgecolor='k')

#specify hue=categorical_variable
sns.boxplot(x='Customer_Segment', y='Profit', hue='Product_Category', data=df)
plt.show()
````

With the help of the graph above, you can see that the **technology segment** is doing really well across all the segments, in pace that **furniture** is incurring losses across all the segments.

You may think that the reason about **Furniture** being generating losses across all the segments is due to the shipping costs. To be sure about this hyphotesis, you can plot derived variables as well, such as shipping cost as percentage of sales amount.

````python
#plot shipping cost as percentage of sales amount
sns.boxplot(x=df['Product_Category'],y=100*df['Shipping_Cost']/df['Sales'])
plt.ylabel('100*(Shipping cost/Sales)')
plt.show()
````

# Plotting Aggregated Values across Categories


## Bar plots - Mean, Median and Count Plots

Bar plots are used to **display aggregated values** of a variable, rather than entire distributions. This is especially usefull when you have a lot of data which is difficult to visualise in a single figure.

For example, say you want to visualise and compare the average Sales across Product Categories. The ``sns.barplot()`` can be used to do that.

````python
#barplot with default statistic=mean
sns.barplot(x='Product_Category', y='Sales', data=df)
plt.show()
````

Note that, **by defaut, seaborn plots the mean value across categories**, though you can plot the count, median, sum, etc. Also, bar plot computes and shows the confidence interval of the mean as well.

````python
#setting the figure size
plt.figure(figsize=(12,6))

#subplot 1: statistic=mean
plt.subplot(1,2,1)
sns.barplot(x='Product_Category', y='Sales', data=df)
plt.title('Average Sales')

#subplot 2: statistic=median
plt.subplot(1,2,2)
sns.barplot(x='Product_Category', y='Sales', data=df, estimator=np.median)
plt.title('Median Sales')

plt.show()
````

The mean and median sales across  the product categories tell diferent stories. This is because of some outliers (extreme values) in the ``Furniture`` category, distorting the value of the mean.

You can add another categorical variable in the plot.


````python
#set figure size for larger figure
plt.figure(num=None, figsize=(12,8), dpi=80, facecolor='w', edgecolor='k')

#especify hue=categorical_variable
sns.barplot(x='Customer_Segment', y='Profit', hue='Product_Category', data=df, estimator=np.median)
plt.show()
````

When you want to visualise having a large number of catgories, is helpful to plot the categories across the y-axis. Let's now drill down into product sub categories

````python
#Plotting categorical variable across the y-axis
plt.figure(figsize=(10,8))
sns.barplot(x='Profit', y='Product_Sub_Category', data=df, estimator=np.median)
plt.show()
````

The plot clearly shows which sub categories are incurring the heaviest losses - Copiers and Fax, Tables, Chairs, and Chairmats are the most loss making categories.

You can also plot the **count of the observations** across categorical variables using ``sns.countplot()``.

````python
#Plotting count across a categorical variable
plt.figure(figsize=(10,8))
sns.countplot(y='Product_Sub_Category', data=df)
plt.show()
````

# Visualising Time Series Data


````python
df = pd.merge(market_df, orders_df, how='inner', on='Ord_id')
df.head()
````

````python
#seeing the type of the df
df.info()
````

Since ``Order_Date`` is a string, we need to convert it into a ``datetime`` object. You can also do that using ``pd.to_datetime()``.

````python
#convert Order_Date to datetime type
df['Order_Date'] = pd.to_datetime(df['Order_Date'])

#Order_Date is now a datetime type
df.info()
````

Now, since on each day, multiple orders were placed, we need to  agregate ``Sales`` using a metric such as mean, median, etc, and then create a time series plot.

We will group by ``Order_Date`` and compute the sum of ``Sales`` on each day.

````python
#aggregating total sales on each day
time_df = df.groupby('Order_Date')['Sales'].sum()
print(time_df.head())

print(type(time_df))
````

Now we can create a time-series plot using ``sns.tsplot()``.

````python
#Time series plot

#figure size
plt.figure(figsize=(16,8))

#tsplot
sns.tsplot(data=time_df)
plt.show()
````

## Using Derived Date Metrics for Visualisation

It is often helpful to use derived variables from date such as month and year and using them to identify hidden patterns.

````python
#Extracting month and year from date

#extracting month
df['month'] = df['Order_Date'].dt.month

#extracting year
df['year'] = df['Order_Date'].dt.year

df.head()
````

````python
# Now we can plot the average sales over month and year
df_time = df.groupby(['year', 'month']).Sales.mean()
df_time.head()
````

````python
plt.figure(figsize=(8,6))

sns.tsplot(df_time)
plt.ylabel('Sales')
plt.xlabel('Time')
````

We can pivot the month to see the sales in a better way:

````python
#Pivoting data using 'month'
year_month = pd.pivot_table(df, values="Sales", index='year', columns='month', aggfunc='mean')
year_month.head()
````

````python
#creating the heatmap
plt.figure(figsize=(12,8))

#heatmap with a color map of choice
sns.heatmap(year_month, cmap='YlGnBu')
plt.show()
````
